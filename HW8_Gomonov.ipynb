Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId

```python
import pandas as pd

ratings = pd.read_csv('ml-latest-small/ratings.csv')

ratings_grouped = ratings.groupby('userId').count()['rating'].reset_index()
ratings_filtered = ratings_grouped[ratings_grouped.rating >= 100]
film_fans_user_ids = ratings_filtered['userId'].tolist()
user_more100 = ratings[ratings['userId'].isin(film_fans_user_ids)]

#ищем время жизни
time_life = user_more100.pivot_table(index=['userId'], values=['timestamp'], aggfunc=[max,min]).reset_index()
time_life.head()

```

```python
time_life['average'] = time_life['max'] - time_life['min']
time_life.head(10)
```
