Задание 2

Создайте диагональную матрицу с элементами от N до 0. Посчитайте сумму ее значений на диагонали.

```python
import numpy as np
N = int(input('Введите количество элементов диагональной матрицы: '))
N_matrix = np.diag(np.arange(N), k=0)
print(N_matrix)
matrix_diag_sum = sum (np.diag(N_matrix))
print("Сумма значений по диагонали равна ", matrix_diag_sum)
```
