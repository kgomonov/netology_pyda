Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
max_rating = ratings.query('rating == 5')['movieId'].value_counts().head(1)

movies = pd.read_csv('movies.csv')

title = movies.query(f'movieId == {max_rating.keys()[0]}').iloc[0]['title']

print(f'Фильму "{title}" было поставленно больше всего оценок 5')
```
