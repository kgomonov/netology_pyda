```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flats = pd.read_csv('flats_moscow.csv')
flats.info()
#проанализируем основные по квартирам в г.Москва
```

```python
flats.head(5)
```

```python
flats.drop(flats.columns[[0]], axis='columns')
```

```python
flats.price.hist(bins=30)
plt.title('Гистограмма распределения стоимости квартир')
```

```python
flats.totsp.hist(bins=20)
plt.title('Гистограмма распределения площади квартир')
```

```python
sns_plot = sns.pairplot(flats[['price', 'totsp', 'livesp','dist']])
sns_plot
#рассмотрим корреляцию между рядом параметров: ценой на квартиру, общей площадью, 
#жилой площадью и расстоянием от центра
#Вывод: расстояние от центра не влияет на стоимость, корреляция 
```

```python
sns.distplot(flats.price)
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
trace0 = go.Scatter(
    x=flats.index,
    y=flats.price,
    name='Price for flats'
)

data = [trace0]
layout = {'title': 'Price for flats'}

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
